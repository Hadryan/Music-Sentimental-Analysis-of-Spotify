```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Reading the csv file into a dataframe
df = pd.read_csv("data.csv", encoding='utf-8')
df= df.dropna()

#Storing all the values in the coloumns into a list
track_id = df['track_id'].values.tolist()
playlist_id = df['playlist_id'].values.tolist()
date_added = df['date_added'].values.tolist()
track_name = df['track_name'].values.tolist()
first_artist = df['first_artist'].values.tolist()
danceability = df['danceability'].values.tolist()
energy = df['energy'].values.tolist()
key = df['key'].values.tolist()
loudness = df['loudness'].values.tolist()
mode = df['mode'].values.tolist()
speechiness = df['speechiness'].values.tolist()
acousticness = df['acousticness'].values.tolist()
instrumentalness = df['instrumentalness'].values.tolist()
liveness = df['liveness'].values.tolist()
valence = df['valence'].values.tolist()
tempo = df['tempo'].values.tolist()
type = df['type'].values.tolist()
id = df['id'].values.tolist()
uri = df['uri'].values.tolist()
track_href = df['track_href'].values.tolist()
analysis_url = df['analysis_url'].values.tolist()
duration_ms = df['duration_ms'].values.tolist()
time_signature = df['time_signature'].values.tolist()
```

```python
#Method used to derive insights based on who the most popular artist is by
# counting the number of songs written by him/her in the dataset
def most_popular_artist():
    res = []
    for i in first_artist:
        if i not in res:
            res.append(i)
    #counts the number of songs written by each artisit in the dataset
    count_list = []
    for i in res:
        count = first_artist.count(i)
        count_list.append(count)
    objects = res
    y_pos = np.arange(len(objects))
    performance = count_list

    # Bar graph used to plot the artist names on the X axis and the number of
    # songs written by them on the Y axis
    plt.bar(y_pos, performance, align='center', alpha=0.5)
    plt.xticks(y_pos, objects, rotation = 90)
    plt.ylabel('Artists')
    plt.title('Artists popularity')
    plt.show()


most_popular_artist()
```

```python
#this method is used to plot a heat map that depicts the correlation between
# each and every feature with the others
def correlation_graph():
    new_df = pd.DataFrame(list(zip(danceability, energy, loudness, acousticness,
                                   instrumentalness,liveness, valence,tempo)),
                          columns=['Danceability', 'Energy', 'Loudness',
                                   'Acousticness', 'Instrumentalness',
                                   'Liveliness', 'Valence', "Tempo"])
    corr = new_df.corr()
    fig, ax = plt.subplots(figsize=(0.5, 0.5))
    ax.matshow(corr)
    plt.xticks(range(len(corr.columns)), corr.columns)
    plt.yticks(range(len(corr.columns)), corr.columns)
    plt.show()

correlation_graph()
```

```python
#scatter plot used to show the relationship between danceability and valence
def scatter_plot():
    ranges = []
    for i in range(0,149):
        ranges.append(0.667*i)
    ranges.append(1)
    #plotting the scatter plot
    plt.scatter(ranges, danceability, color='r', label = "Danceability")
    plt.scatter(ranges, valence, color='g', label = "Valence")
    plt.xlabel = 'Danceability'
    plt.ylabel = 'Valence'
    plt.legend(loc = "upper right")
    plt.show()

scatter_plot()
```

```python
# scatter_plot()
#this method is used to plot a bar graph that shows the mean values of the audio
#  features
def mean_graph():
    #calculating the arithmetic mean of all the features
    danceability_mean = np.mean(danceability)
    energy_mean = np.mean(energy)
    mode_mean = np.mean(mode)
    speechiness_mean = np.mean(speechiness)
    acousticness_mean = np.mean(acousticness)
    instrumentalness_mean = np.mean(instrumentalness)
    liveness_mean = np.mean(liveness)
    valence_mean = np.mean(valence)

    #plotting the ba graph
    mean_list = [danceability_mean, energy_mean, mode_mean, speechiness_mean, acousticness_mean, instrumentalness_mean, liveness_mean, valence_mean]
    objects = ('Danceability', 'Energy', 'Mode', 'Speechiness', 'Acousticness', 'Instrumentalness', 'Liveness', 'Valence')
    y_pos = np.arange(len(objects))
    performance = mean_list
    plt.bar(y_pos, performance, align='center', alpha=0.5)
    plt.xticks(y_pos, objects)
    plt.ylabel('Values')                                                                      
    plt.title('Features')
    plt.show()

mean_graph()
```
